```python
import datetime
import pandas as pd
import requests
import matplotlib as mpl
import matplotlib.pyplot as plt
import json

mpl.rcParams['figure.figsize'] = (12,6)
mpl.rcParams['axes.grid'] = True
%matplotlib inline
```

```python
coindeskURL = 'https://api.coindesk.com/v1/bpi/historical/close.json?'

start = datetime.date(2017, 1 ,1)
end = datetime.date(2022, 8, 12)

url = f'{coindeskURL}start={start:%Y-%m-%d}&end={end:%Y-%m-%d}'

result = requests.get(url)
result.content
```

```python
data = json.loads(result.content)
data = pd.DataFrame(data)
data.head()
```

```python
from datetime import datetime
df = data[['bpi']]
index = df.index
index = pd.Series(index)
index = index[:-2]
index = index.apply(lambda x:datetime.strptime(x, '%Y-%m-%d'))
df = df.iloc[:-2]
df.index = index
```

```python
mpl.rcParams['figure.figsize'] = (14,7)
df.plot()
```

#### Resampling 
- We do resampling in order to reduce data points being fed to model by keeping the pattern of the data

```python
df_c = df.copy()
```

```python
# Aggregating the data on the quarterly basis

df.resample('Q').mean()
```

```python
df.resample('Q').mean().plot()
```

```python
# Resampling the data on monthly level

df.resample('M').mean().plot()
```

#### Making the data stationary

```python
print(df[:5])
```

```python
print(df[:5].diff())
```

```python
print(df[:5].diff(2))
```

```python
pd.concat(([df, df.diff(), df.diff(2)]), axis = 1).plot()
```

- From the above graph we can see trends have been removed and data has been made stationary and there are some seasonality/periodity

```python
# Data will be shifted based on the index
df[:5].tshift(3)
```

```python
df_c['Shift'] = df_c['bpi'] - df_c['bpi'].shift(2)
df_c['Diff'] = df_c['bpi'].diff(2)
df_c.head()
```

```python
print(df_c['bpi'].shift(2))
```

```python
df_c['bpi'].rolling(window = 10).mean().plot()
```

### In the exponential weighted average, we give the different weights for previous time stamp like if we define the alpha then alpha with higher alpha denotes more weightage to recent data
### But in rolling window we give equal weightage 

```python
pd.concat([df_c['bpi'], df_c['bpi'].rolling(window=100).mean(), df_c['bpi'].ewm(span=100).mean()], axis=1).plot()
```

```python
df_c['Week resampled'] = df_c['bpi'].resample('W').mean() 
```
